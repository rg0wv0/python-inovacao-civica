### Acesse a api da camara, e recupere os reembolsos do ano de 2018

```python
import requests

# fazendo o download do conteúdo do arquivo
url = "http://www.camara.leg.br/cotas/Ano-2018.csv.zip"

r = requests.get(url)

file = open("Ano-2018.csv.zip", "wb")
file.write(r.content)
file.close()

from zipfile import ZipFile
zip_file = ZipFile("Ano-2018.csv.zip", 'r')

zip_file.extractall("reimbursements-2018")
zip_file.close()
```

### Converta a coluna `datEmissao` para `datetime`, e descubra quantos reembolsos aconteram no primeiro semestre

```python
df_reimbursements_2018["datEmissao"] = pd.to_datetime(
    df_reimbursements_2018.datEmissao, 
    format='%Y-%m-%d'
) 

df_reimbursements_2018.set_index("datEmissao").loc["2018-01-01":"2018-06-30"].sort_values(by=['datEmissao','sgPartido'], ascending=True)
```

```python
df_reimbursements_2018.set_index("datEmissao").loc["2018-01-01":"2018-06-30"].info()
df_reimbursements_2018.set_index("datEmissao").loc["2018-01-01":"2018-06-30"].shape
```

### Limpe a coluna `txtCNPJCPF`: ela deve conter somente dígitos numéricos

```python
# vamos trocar qualquer caracter que não seja um dígito número por '', essa operação remove os dígitos
#df_reimbursements["txtCNPJCPF"] = df_reimbursements...
df_reimbursements_2018["txtCNPJCPF"] = df_reimbursements_2018["txtCNPJCPF"].str.replace(r'\D', '', regex=True)
df_reimbursements_2018.head()
```
